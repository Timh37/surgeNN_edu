```python
import xarray as xr
import dask
import os
import gcsfs
fs = gcsfs.GCSFileSystem() #list stores, stripp zarr from filename, load 
```

Move locally stored files to GCS

```python
in_path = '/home/jovyan/surgeNN/results/nns/corrected_splitting/performance/convlstm/'
cloud_path = 'gs://leap-persistent/timh37/surgeNN_output/nns/performance_modified_v2/convlstm/'
try:
    fs.ls(cloud_path)
except:
    pass
```

```python
for file in os.listdir(in_path):
    if file.startswith('.') == False:
        print(file)
        fn = os.path.join(in_path,file)
        ds  = xr.open_dataset(fn)
        for k in ['p','tg','split']:
            ds[k] = ds[k].astype('str')
        ds.to_zarr(os.path.join(cloud_path,file.replace('.nc','.zarr').replace('_ndeg5','')),mode='w',zarr_version=2)
fs.ls(cloud_path)
```
